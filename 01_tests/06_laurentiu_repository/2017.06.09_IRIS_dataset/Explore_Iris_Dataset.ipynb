```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

headers = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"] 
iris = pd.read_csv("iris.csv", names=headers)

print("Total number of observations in IRIS dataset = {0}; nr_cols = {1}\n".format(len(iris), len(iris.columns)))
print("Top 5 observations in IRIS dataset:")
print(iris.head(5))
```

```python
# Count, mean, std deviation. min, max, quartiles
print(iris.describe())
```

```python
# The unique values and their frequency of variable 'species'
print(iris.species.value_counts())
```

```python
# Missing values
print(iris.isnull().sum()) # no missing values
```

```python
def percConvert(ser):
  return ser/float(ser[-1])

""" Using crosstab and boxplot in order to see how species 
    are related to sepal_{length, width} and to petal_{length, width} """

iris.boxplot(column='sepal_length', by='species')
#pd.crosstab(iris["sepal_length"],iris["species"],margins=True).apply(percConvert, axis=1)
```

```python
iris.boxplot(column='sepal_width', by='species')
#pd.crosstab(iris["sepal_width"],iris["species"],margins=True).apply(percConvert, axis=1)
```

```python
iris.boxplot(column='petal_length', by='species')
#pd.crosstab(iris["petal_length"],iris["species"],margins=True).apply(percConvert, axis=1)
```

```python
iris.boxplot(column='petal_width', by='species')
#pd.crosstab(iris["petal_width"],iris["species"],margins=True).apply(percConvert, axis=1)
```

```python
# Histograms related to boxplots
iris.hist(column='sepal_length', by='species', bins=15)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
iris.hist(column='petal_length', by='species', bins=15, ax=axes)
plt.suptitle('Petal Length histogram', x=0.5, y=1.05, ha='center', fontsize='xx-large')
fig.text(0.5, 0.04, 'Petal Length', ha='center')
fig.text(0.04, 0.5, 'Frequency', va='center', rotation='vertical')
```

```python
plt.xlabel('Sepal width')
plt.ylabel('Frequency')
iris.sepal_width.hist(bins=15)
```

```python
# Perform log transformation of sepal_width to make it closer to normal
iris['sepal_width_log']= np.log(iris['sepal_width'])
plt.xlabel('Sepal width log')
plt.ylabel('Frequency')
iris.sepal_width_log.hist(bins=15)
```
